```python
df=pd.read_csv(r'C:\Users\sai\Desktop\PARITOSH\google data analst\Case Study Capstone Project\movies.csv')
```

```python
#Lets look at data
df.head()
```

```python
#lets see if there is any missing data
for cols in df.columns:
    pct_missing=np.mean(df[cols].isnull())
    print('{} - {}%'.format(cols,pct_missing))
```

```python
# data types for our coloums

df.dtypes
```

```python
#finding null values
df.isnull().sum()
```

```python
#removing missing values
df.dropna(inplace=True)
```

```python
#changing data type
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
df['votes']= df['votes'].astype('int64')
df['score']= df['gross'].astype('int64')
df['runtime']= df['runtime'].astype('int64')
```

```python
df
```

```python
# creating a Year column
df['Year']=df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
#df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show()
```

```python
# Plot budget Vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#High correlation betwwen budget and gross
```

```python
correlation_matrix=df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation_matrix for Numeric Feature')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show()
```

```python
# looks at company
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation_matrix for Numeric Feature')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show()
```
